```python
import numpy as np
ar2=np.random.randint(1,30,size=(3,6))
jh=ar2.reshape(9,2)
#transpose
JH=jh.T
#sum along columns
hg=JH.sum(axis=1)
#sum along rows
gf=hg.sum(axis=0)
mean=gf/18
fd=jh[jh>mean]
print(fd)

# Create a Poisson distribution array of length 20 with lambda=5
```

```python
poisson_array = np.random.poisson(5, size=20)

# Calculate the mean of the Poisson array
poisson_mean = np.mean(poisson_array)

# Center the array around its mean
centered_poisson_array = poisson_array - poisson_mean

# Calculate the standard deviation of the centered array
poisson_std = np.std(centered_poisson_array)

# Normalize the centered array by dividing by its standard deviation
normalized_poisson_array = centered_poisson_array / poisson_std

print("\nOriginal Poisson Array:", poisson_array)
print("Centered Poisson Array:", centered_poisson_array)
print("Normalized Poisson Array:", normalized_poisson_array)
```
