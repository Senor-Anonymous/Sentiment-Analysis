```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Create a numpy array for the normal distribution
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Create a numpy array for the beta distribution
beta_data = np.random.beta(a=2, b=5, size=1000)

# Create 2 subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the normal distribution histogram
axes[0].hist(normal_data, bins=20, density=True, alpha=0.7, color='skyblue', edgecolor='black')
axes[0].set_title('Normal Distribution (Mean=0, Std=1)')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Density')

# Plot the beta distribution histogram
axes[1].hist(beta_data, bins=20, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
axes[1].set_title('Beta Distribution (a=2, b=5)')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a numpy array of 1000 random values from a normal distribution
normal_data_new = np.random.normal(loc=0, scale=1, size=1000)

# Calculate mean, 25th percentile, and 75th percentile
mean_val = np.mean(normal_data_new)
p25_val = np.percentile(normal_data_new, 25)
p75_val = np.percentile(normal_data_new, 75)

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(normal_data_new, bins=20, density=True, alpha=0.7, color='lightpink', edgecolor='purple', label='Normal Distribution')

# Add vertical lines
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(p25_val, color='blue', linestyle='dotted', linewidth=2, label=f'25th Percentile: {p25_val:.2f}')
plt.axvline(p75_val, color='green', linestyle='dotted', linewidth=2, label=f'75th Percentile: {p75_val:.2f}')

# Annotate the plot
plt.text(mean_val + 0.1, plt.gca().get_ylim()[1] * 0.9, f'Mean: {mean_val:.2f}', color='red', verticalalignment='top')
plt.text(p25_val + 0.1, plt.gca().get_ylim()[1] * 0.8, f'25th: {p25_val:.2f}', color='blue', verticalalignment='top')
plt.text(p75_val + 0.1, plt.gca().get_ylim()[1] * 0.7, f'75th: {p75_val:.2f}', color='green', verticalalignment='top')

# Customize visualization
plt.title('Histogram of Normal Distribution with Mean and Percentiles', color='darkorchid')
plt.xlabel('Value', color='darkmagenta')
plt.ylabel('Density', color='darkmagenta')
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.show()
```
